### cta_025D

```python
### define basic path
asset_ver_dict = {'cta':{'025D':{'asset':'cta025D','ver':'v2'}}}
asset = 'cta'
in_freq = '025D'
out_freq = '025D'
input_format = 'h5'
output_format = 'h5'
basic_path = get_data_dir(asset_ver_dict[asset][in_freq]['asset'],asset_ver_dict[asset][in_freq]['ver'],data_format=input_format)
basic_path
```

```python
### define factorPools, and label list
train_start = 2010
test_year_list = [2019,2020,2021,2022,2023,2024]
y_list = ["y4", "y20", "y40", "y80"]
label_path = '/cpfs/dss/dev/cta/project/flap01/p9/factor_set/binary/jchen/priceMain_set/025D/cta/autoRedig/saved/label_volAdj/'
factors_pathmap = '/cpfs/dss/dev/cta/project/flap01/p9/factor_set/binary/jchen/priceMain_set/025D/cta/autoRedig/autopriceMain_first_step_volDealt_pathmap.csv'
selectF_infos = '/cpfs/dss/dev/cta/project/flap01/p9/factor_set/binary/jchen/priceMain_set/025D/cta/autoRedig/filter_json/json_res/first_step_volDealt_rnd2_corr08/autopriceMain_first_step_volDealt_thres_top_150_trunc_icts/'
factor_pathInfos = pd.read_csv(factors_pathmap)
```

```python
### example to show based on train_start and train_end to merged label and factors(based on select json)
for y in y_list[:1]:
    for test_year in test_year_list[:1]:
        print(y, test_year)
        label = merge_factor(basic_path, [label_path+y+'.'+input_format], input_format=input_format,start_year=train_start, end_year=test_year-1, verbose=0)
        label = label.rename(columns = {'date':'time'})
        selectF_fname = selectF_infos+y+'/'+str(test_year)[2:]+'.json'
        with open(selectF_fname, 'r') as file:
            selectF_list = [i.split('.')[2] for i in json.loads(file.read())['factor_list']]
        selectF_paths = sorted(factor_pathInfos[factor_pathInfos['fname'].isin(selectF_list)]['bin_path'])
        factors = merge_factor(basic_path, selectF_paths,n_jobs=40, input_format=input_format,start_year=train_start, end_year=test_year-1)
```
